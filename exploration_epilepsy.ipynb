```python
import pandas as pd
import os
import numpy as np
import mne

sub1 = '/home/mat/projects/def-kjerbi/cocolab_data/EEG_ADHD_epilepsy_psychostimulants/BIDS/sub-0039/ses-01/eeg/sub-0039_ses-01_task-RESTING_run-01_eeg.vhdr'
#load data
raw = mne.io.read_raw_brainvision(sub1, preload=True)
# basic info : Number of channels: 40, channel names,  sampling frequency: 200 Hz, channel types: eeg misc, measurement date: unspecified h/l pass 
print(raw.info)
# plotting raw signals 
raw.plot(n_channels=35, duration=5, scalings='auto')

```

Power Spectral Density PSD: desscribes how the power of a signal is distributed across different frequencies. used to:
- Identify dominant brainwave rhythms
- Detect abnormalities
- Compare brain states


```python
raw.plot_psd(fmax=100)
# looks very stable
```

```python
events, event_id = mne.events_from_annotations(raw)
print(events[:10])  # show first 10 events
#print(event_id)     # mapping of event labels to numbers
```

```python
mne.viz.plot_events(events, sfreq=raw.info['sfreq'])
```
